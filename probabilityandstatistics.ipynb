```python
# Given values from the table
A_and_B = 30   # Number of occurrences of A and B
A_and_not_B = 20   # Number of occurrences of A and not B
not_A_and_B = 50   # Number of occurrences of not A and B
not_A_and_not_B = 100  # Number of occurrences of not A and not B

# Total number of occurrences for A, B, and AB
P_A_and_B = A_and_B
P_B = A_and_B + not_A_and_B  # Total occurrences of B
P_A = A_and_B + A_and_not_B  # Total occurrences of A

# Conditional probability P(A|B) = P(A  B) / P(B)
P_A_given_B = P_A_and_B / P_B

# Conditional probability P(B|A) = P(A  B) / P(A)
P_B_given_A = P_A_and_B / P_A

# Print results
print(f"P(A|B) = {P_A_given_B:.3f}")
print(f"P(B|A) = {P_B_given_A:.3f}")
```

```python
# Given values
P_Spam = 0.2
P_Not_Spam = 0.8
P_Keyword_given_Spam = 0.7
P_Keyword_given_Not_Spam = 0.1

# Step 1: Calculate P(Keyword)
P_Keyword = (P_Keyword_given_Spam * P_Spam) + (P_Keyword_given_Not_Spam * P_Not_Spam)

# Step 2: Apply Bayes' Theorem
P_Spam_given_Keyword = (P_Keyword_given_Spam * P_Spam) / P_Keyword

# Display the result
print(f"P(Spam | Keyword) = {P_Spam_given_Keyword:.3f}")
```

```python
# Function to remove outliers
def remove_outliers(dataframe):
    # Compute Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = dataframe.quantile(0.25)
    Q3 = dataframe.quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Define the bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = (dataframe < lower_bound) | (dataframe > upper_bound)

    # Remove rows with outliers
    cleaned_dataframe = dataframe[~outliers.any(axis=1)]

    return cleaned_dataframe
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_diabetes

# Load the Diabetes dataset
diabetes = load_diabetes()

# Convert the data into a DataFrame and select features up to 'bp'
diabetes_df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
diabetes_df = diabetes_df[['age', 'sex', 'bmi', 'bp']]

# Function to remove outliers
def remove_outliers(dataframe):
    # Compute Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = dataframe.quantile(0.25)
    Q3 = dataframe.quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Define the bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = (dataframe < lower_bound) | (dataframe > upper_bound)

    # Remove rows with outliers
    cleaned_dataframe = dataframe[~outliers.any(axis=1)]

    return cleaned_dataframe

# Remove outliers
cleaned_diabetes_df = remove_outliers(diabetes_df)

# Plot boxplots of features before and after outlier removal
def plot_boxplots_before_after(dataframe, cleaned_dataframe):
    # Create a 1x2 subplot layout
    plt.figure(figsize=(14, 6))
    
    # Boxplot before outlier removal
    plt.subplot(1, 2, 1)
    sns.boxplot(data=dataframe)
    plt.title("Boxplot Before Outlier Removal")
    
    # Boxplot after outlier removal
    plt.subplot(1, 2, 2)
    sns.boxplot(data=cleaned_dataframe)
    plt.title("Boxplot After Outlier Removal")
    
    # Display the plots
    plt.tight_layout()
    plt.show()

# Plotting boxplots for the Diabetes dataset
plot_boxplots_before_after(diabetes_df, cleaned_diabetes_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Function to plot histograms of features before and after normalization/standardization
def plot_histograms_before_after(dataframe):
    # Select a few features to plot
    features_to_plot = ['age', 'bmi', 'bp']
    
    # Initialize the scalers
    min_max_scaler = MinMaxScaler()
    standard_scaler = StandardScaler()

    # Apply normalization (Min-Max Scaling)
    normalized_df = dataframe.copy()
    normalized_df[features_to_plot] = min_max_scaler.fit_transform(dataframe[features_to_plot])

    # Apply standardization (Z-score Scaling)
    standardized_df = dataframe.copy()
    standardized_df[features_to_plot] = standard_scaler.fit_transform(dataframe[features_to_plot])

    # Create a 3x2 subplot layout
    plt.figure(figsize=(15, 10))
    
    # Before normalization/standardization
    for i, feature in enumerate(features_to_plot, 1):
        plt.subplot(3, 2, i)
        sns.histplot(dataframe[feature], kde=True)
        plt.title(f"Original {feature}")
    
    # After normalization
    for i, feature in enumerate(features_to_plot, 4):
        plt.subplot(3, 2, i)
        sns.histplot(normalized_df[feature], kde=True)
        plt.title(f"Normalized {feature}")
    
    # After standardization
    for i, feature in enumerate(features_to_plot, 7):
        plt.subplot(3, 2, i-3)  # Corrected subplot numbering
        sns.histplot(standardized_df[feature], kde=True)
        plt.title(f"Standardized {feature}")
    
    # Display the plots
    plt.tight_layout()
    plt.show()

# Example usage: Plotting histograms for the Diabetes dataset
plot_histograms_before_after(diabetes_df)
```
